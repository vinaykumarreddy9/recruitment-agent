```python
from langgraph.graph import END, StateGraph, MessagesState
```

```python
class State(MessagesState):
    pass
```

```python
def supervisor_agent(State):
    pass
def router_node(State):
    pass
def intent_agent(State):
    pass
def jd_agent(State):
    pass
def questioning_agent(State):
    pass
```

```python
app = StateGraph(State)

app.add_node("supervisor_agent", supervisor_agent)
app.add_node("intent_agent", intent_agent)
app.add_node("jd_agent", jd_agent)
app.add_node("questioning_agent", questioning_agent)

app.set_entry_point("supervisor_agent")
app.add_conditional_edges("supervisor_agent", router_node, {
    "invoke_intent": "intent_agent",
    "invoke_jd": "jd_agent",
    "invoke_questions": "questioning_agent",
    "END": END,
})

app.add_edge("intent_agent","supervisor_agent")
app.add_edge("jd_agent","supervisor_agent")
app.add_edge("questioning_agent","supervisor_agent")

graph= app.compile()
```
